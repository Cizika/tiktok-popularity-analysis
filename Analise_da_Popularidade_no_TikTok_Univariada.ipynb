## 1. Introdução

Nos últimos anos, plataformas de vídeos curtos, como o TikTok e o Instagram Reels, conquistaram o mercado de aplicativos, tornando-se exponencialmente populares entre os usuários digitais. Segundo relatório da especialista em marketing digital Sensor Tower, o TikTok superou a marca de 3 bilhões de downloads em 2021, dispondo de um bilhão de usuários ativos mensais em todo o mundo. Essa rápida ascensão se deve, principalmente, à capacidade dos usuários em gravar, compartilhar e consumir vídeos curtos de diferentes atividades, como cantar, dançar e dublar. Além disso, o algoritmo de recomendação é altamente eficiente em exibir conteúdos personalizados, mantendo os usuários engajados com base em seus interesses e comportamentos.

Nesse contexto, a música tornou-se uma importante parte da experiência, surgindo assim um novo modo de exposição e promoção da arte. Isso porque canções que viralizam nessas plataformas experimentam um aumento de vendas e visualizações no streaming (Nielsen Music/MRC, 2022), impulsionando a carreira de artistas que antes eram pouco conhecidos pelo público. De acordo com relatórios da própria plataforma, 75% dos usuários descobriram novos artistas através do TikTok, enquanto 63% escutaram novas músicas pela primeira vez na plataforma. 

Esse fenômeno alterou o modo como as músicas são produzidas e pensadas, pressionando a indústria musical a encontrar formas de tornar suas músicas atrativas aos algoritmos de recomendação. Nessa conjuntura, o projeto aqui proposto tem como objetivo estudar o que torna uma música popular no TikTok, conduzindo uma análise das características e informações acerca de músicas para predizer sua popularidade nas plataformas.

Sendo assim, essa análise propõe estudos a fim de obter conclusões acerca da popularidade de uma música por meio de seus atributos, aplicando-se algoritmos de aprendizado de máquina. Assim, poder-se-á classificar uma música como popular ou não, além de obter através de modelos de regressão quais características fazem com que uma música seja mais popular do que outra.

Tal análise faz-se altamente necessária no contexto da indústria musical atual, já que gravadoras sempre se empenharam em descobrir padrões que façam com que seus artistas produzam sucessos e, por conseguinte, façam-nas lucrar mais. Dessa maneira, ao determinar se uma música é popular, ou ainda, quais atributos a fariam tornar-se um viral, é possível que os artistas possam trabalhar de maneira a criar hits instantâneos.

### Dicionário de dados

* **Popularidade da Música**: Variável resposta do problema, em que 0 é pouco popular e 100 é muito popular;
* **Nome da música**;
* **Nome do artista**;
* **Duração**: Duração da música, dada em milissegundos;
* **Data de Lançamento**: Data em que a música foi lançada no formato YYYY-MM-DD;
* **Danceability**: Descreve o quão dançável a música é com base na combinação de tempo, ritmo, força da batida e regularidade geral. (0 a 1);
* **Energia**: Medida (de 0 a 1) que representa a percepção de intensidade e velocidade da música. Relacionada ao timbre, cadência, entropia e volume;
* **Loudness**: Volume médio da música em decibels (dB);
* **Acousticness**: Confiança de 0 a 1 de que a música seja acústica;
* **Mode**: Presença de notas maiores ou menores na música. Notas maiores são representada por 1 e notas menores por 0;
* **Key**: Tom/afinação da música. Cada valor representa uma classe de notas, em que C=0, C#/Db = 1, D = 2 e assim sucessivamente;
* **Liveness**: Detecta a presença de audiência na gravação. Valores maiores que 0.8 indicam alta probabilidade da música ser ao vivo;
* **Speechiness**: Detecta a presença de palavras na música. Valores próximos de 1 indicam gravações com mais discurso, como poesia ou rap, enquanto valores próximos de 0;
* **Instrumentalness**: Quantifica a presença de vocais na música. Quanto mais próximo de 1, maior a probabilidade de que não haja conteúdo vocal na música;
* **Tempo**: A média do passo/ritmo da música, dada em batidas por minuto (BPM);
* **Valence**: Relacionada à positividade transmitida pela música. Canções alegres e eufóricas possuem valor próximo de 1, enquanto canções tristes e depressivas se aproximam do 0.

```python
testes = []

for variavel in variaveis_numericas:
    resultados = []
    for dist in distribuicoes_continuas:
        args = dist.fit(dados_new[variavel])
        ks = stats.kstest(dados_new[variavel], dist.cdf, args)
        resultados.append([dist.name, dist(*args), ks])
    testes.append((variavel, sorted(resultados, key=lambda res: res[2].pvalue, reverse=True)))
```

```python
# Analisar todos os resultados do teste KS sobre as distribuições selecionadas
dicionario_resultados = {
    variavel: [
        coluna
        for registro in sorted([
            (distribuicao, rank, ks_result.pvalue)
            for rank, (distribuicao, rv, ks_result) in enumerate(resultados, start=1)
        ], key=lambda row: list(map(lambda d: d.name, distribuicoes_continuas)).index(row[0]))
        for coluna in registro[1:]
    ]
    for variavel, resultados in testes
}

colunas = pd.MultiIndex.from_product([
    list(map(lambda d: d.name, distribuicoes_continuas)),
    ['posição', 'valor-p']
])


resultados = pd.DataFrame.from_dict(dicionario_resultados, orient='index', columns=colunas)
```

```python
melhor_distribuicao = pd.DataFrame.from_dict({
    variavel: [distribuicao, f"{ks.pvalue:.4%}"]
    for variavel, resultados in testes
    for distribuicao, _rv, ks in resultados[:1]
}, orient='index', columns=('distribuição', 'valor-p'))
melhor_distribuicao
```

```python
MOSTRAR_RESULTADOS: int = 3

fig, axes = plt.subplots(len(testes), MOSTRAR_RESULTADOS, figsize=(15, 70), sharey='row')

for (variavel, resultados), axes_row in zip(testes, axes):
    for (distribuicao, rv, ks), ax in zip(resultados[:MOSTRAR_RESULTADOS], axes_row):
        points = np.linspace(rv.ppf(0.01), rv.ppf(0.99), 100)
        pdf = rv.pdf(points)
        if variavel != 'instrumentalness':
            sns.histplot(dados_new[variavel], stat='density', label='data', color=palette[0], ax=ax)
        else:
            sns.histplot(dados_new[variavel], stat='density', label='data', bins=40, color=palette[0], ax=ax)
            ax.set_yscale('log')
        sns.lineplot(x=points, y=pdf, color=palette[1], label=f'{distribuicao} (p-value: {ks.pvalue:.4f})', ax=ax)
        ax.set_ylabel(f"{variavel} | density")
        ax.set_title(distribuicao)
```
